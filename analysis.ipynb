```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
data = pd.read_csv('clean_data.csv')
data
```

```python
# Drop index column
data = data.drop('index', axis=1)
```

```python
# Columns
data.columns
```

```python
# Set style
sns.set_style('darkgrid')
```

```python
# Food category
plt.figure(figsize=(20,20))
sns.countplot(data=data, y='food_category', palette='tab20b')
plt.title('Food category', fontdict={'fontsize':20})
plt.show()

# Food category by cost
plt.figure(figsize=(20,10))
ax = sns.boxplot(data=data, x='food_category', y='cost', palette='tab20b')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.title('Food category by Cost', fontdict={'fontsize':20})
plt.show()
```
